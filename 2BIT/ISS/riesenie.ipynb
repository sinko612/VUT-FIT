```python
import numpy as np
import soundfile as sf
import matplotlib.pyplot as plt
import statsmodels.api as sm
import scipy.signal as sc

TONE1 = 35
TONE2 = 46
TONE3 = 101

#nacitanie vstupnych tonov
MIDIFROM = 24
MIDITO = 108
SKIP_SEC = 0.25
HOWMUCH_SEC = 0.5
WHOLETONE_SEC = 2
howmanytones = MIDITO - MIDIFROM + 1
tones = np.arange(MIDIFROM, MIDITO+1)
s, Fs = sf.read('klavir.wav')
N = int(Fs * HOWMUCH_SEC)
Nwholetone = int(Fs * WHOLETONE_SEC)
xall = np.zeros((MIDITO+1, N))
samplefrom = int(SKIP_SEC * Fs)
sampleto = samplefrom + N
for tone in tones:
    x = s[samplefrom:sampleto]
    x = x - np.mean(x)
    xall[tone,:] = x
    samplefrom += Nwholetone
    sampleto += Nwholetone
```

```python
#4.1
sf.write('a_orig.wav',xall[TONE1],Fs)
plt.figure(figsize=(8, 7), dpi=80)
plt.subplot(211)
plt.title("Tn MIDI 35 - 3 peridy")
plt.xlabel("as [ms]")
plt.ylabel("Amplitda")
plt.grid()
plt.xlim(-2, 50)
plt.ylim(-0.2, 0.2)
perioda = (1/61.74)
x = np.arange(0,3*perioda*Fs,step=1/Fs)*1000
end = int(3*perioda*Fs)
plt.plot(x[:end],xall[TONE1][:end])
plt.subplot(212)
plt.title("Tn MIDI 35 - spektrum")
plt.xlabel("Frekvencia [Hz]")
plt.ylabel("log PSD [dB]")
plt.xlim(-1000, 25000)
plt.ylim(-15, 15)
DFT = np.fft.fft(xall[TONE1])
DFT = DFT * np.conj(DFT)
DFT = DFT[:DFT.size // 2]
DFT = np.real(DFT) 
DFT = np.log(1e-5+DFT)
x = np.arange(DFT.size) * (Fs / xall[TONE1].size)
plt.plot(x,DFT)
plt.grid()
plt.tight_layout()
plt.show()

sf.write('b_orig.wav',xall[TONE2],Fs)
plt.figure(figsize=(8, 7), dpi=80)
plt.subplot(211)
plt.title("Tn MIDI 46 - 3 peridy")
plt.xlabel("as [ms]")
plt.ylabel("Amplitda")
plt.grid()
plt.xlim(-2, 27)
plt.ylim(-0.2, 0.2)
perioda = (1/116.54)
x = np.arange(0,3*perioda*Fs,step=1/Fs)*1000
end = int(3*perioda*Fs)
plt.plot(x[:end],xall[TONE2][:end])
plt.subplot(212)
plt.title("Tn MIDI 46 - spektrum")
plt.xlabel("Frekvencia [Hz]")
plt.ylabel("log PSD [dB]")
plt.xlim(-1000, 25000)
plt.ylim(-15, 15)
DFT = np.fft.fft(xall[TONE2])
DFT = DFT * np.conj(DFT)
DFT = DFT[:DFT.size // 2]
DFT = np.real(DFT) 
DFT = np.log(1e-5+DFT)
x = np.arange(DFT.size) * (Fs / xall[TONE2].size)
plt.plot(x,DFT)
plt.grid()
plt.tight_layout()
plt.show()

sf.write('c_orig.wav',xall[TONE3],Fs)
plt.figure(figsize=(8, 7), dpi=80)
plt.subplot(211)
plt.title("Tn MIDI 101 - 3 peridy")
plt.xlabel("as [ms]")
plt.ylabel("Amplitda")
plt.grid()
plt.xlim(-0.05, 1.1)
plt.ylim(-0.003, 0.004)
perioda = (1/2793.83)
x = np.arange(0,3*perioda*Fs,step=1/Fs)*1000
end = int(3*perioda*Fs)
plt.plot(x[:end],xall[TONE3][:end])
plt.subplot(212)
plt.title("Tn MIDI 101 - spektrum")
plt.xlabel("Frekvencia [Hz]")
plt.ylabel("log PSD [dB]")
plt.xlim(-1000, 25000)
plt.ylim(-15, 10)
DFT = np.fft.fft(xall[TONE3])
DFT = DFT * np.conj(DFT)
DFT = DFT[:DFT.size // 2]
DFT = np.real(DFT) 
DFT = np.log(1e-5+DFT)
x = np.arange(DFT.size) * (Fs / xall[TONE3].size)
plt.plot(x,DFT)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
#4.2
for tone in tones:
    sample = xall[tone]
    if(tone < 56):
        correlation = np.correlate(sample, sample, mode='full')
        correlation = correlation[len(correlation)//2:]
        d = np.diff(correlation)
        start = np.nonzero(d > 0)[0][0]
        peak = np.argmax(correlation[start:]) + start
        if tone == TONE1:
            plt.subplot(212)
            plt.title("Tn MIDI 35 - autokorelcia")
            plt.xlabel("Vzorky")
            plt.ylabel("Korelan koeficient")
            plt.xlim(-1000, 25000)
            plt.ylim(-85, 125)
            plt.vlines(x=peak, ymin=-85, ymax=125, colors='red', ls='--', lw=2, label='vline_multiple - full height')
            plt.plot(correlation)
            plt.tight_layout()
            plt.grid()
            plt.show()
        elif tone == TONE2:
            plt.subplot(212)
            plt.title("Tn MIDI 46 - autokorelcia")
            plt.xlabel("Vzorky")
            plt.ylabel("Korelan koeficient")
            plt.vlines(x=peak, ymin=-20, ymax=30, colors='red', ls='--', lw=2, label='vline_multiple - full height')
            plt.xlim(-1000, 25000)
            plt.ylim(-20, 30)
            plt.plot(correlation)
            plt.tight_layout()
            plt.grid()
            plt.show()
        print(tone,Fs / peak) 
    elif (tone >= 56):
        fft = np.fft.fft(sample)
        f = np.fft.fftfreq(len(sample))
        k_peak = np.argmax(np.abs(fft))
        f_peak = f[k_peak]
        if tone == TONE3:
            plt.subplot(212)
            plt.title("Tn MIDI 101 - spektrum")
            plt.xlabel("Frekvencia [Hz]")
            plt.ylabel("log PSD [dB]")
            plt.xlim(-1000, 25000)
            plt.ylim(-15, 10)
            plt.vlines(x=abs(f_peak*Fs), ymin=-15, ymax=10, colors='red', ls='--', lw=2, label='vline_multiple - full height')
            x = np.arange(fft.size) * (Fs / xall[tone].size)
            plt.plot(x[:x.size // 2],DFT)
            plt.tight_layout()
            plt.grid()
            plt.show()
        print(tone,abs(f_peak*Fs))  
        
```

```python
#4.3
for tone in tones:
    sample = xall[tone]
    if(tone < 56):
        correlation = np.correlate(sample, sample, mode='full')
        correlation = correlation[len(correlation)//2:]
        d = np.diff(correlation)
        start = np.nonzero(d > 0)[0][0]
        peak = np.argmax(correlation[start:]) + start
        N = len(xall[tone])
        n = np.arange(0,N)
        X = correlation
        FREQRANGE = 50
        if tone < 35:
            FREQRANGE = FREQRANGE//2
        FREQPOINTS = 200
        
        kall = np.arange(0,int(N/2) +1)
        Xmag = np.abs(X[kall])
        Xphase = np.angle(X[kall])
        f = kall / N * Fs
        if tone == TONE1 or tone == TONE2:
            plt.plot(f[:f.size // 8],Xmag[:Xmag.size // 8])
            plt.gca().set_ylabel('$|X[k]|$')

         
        fmax = Fs/peak
        Xmax = np.max(Xmag)
        ffrom = fmax-FREQRANGE
        fto = fmax+FREQRANGE
        if tone == TONE1 or tone == TONE2:
            plt.bar(ffrom,Xmax,fto-ffrom,linewidth=1,align='edge',color='orange',alpha=0.4) 
        fsweep = np.linspace(fmax-FREQRANGE, fmax+FREQRANGE,FREQPOINTS)

        A = np.zeros([FREQPOINTS, N],dtype=complex)   
        for k in np.arange(0,FREQPOINTS):
            A[k,:] = np.exp(-1j * 2 * np.pi * fsweep[k] / Fs * n)
        Xdtft = np.matmul(A,xall[tone].T)
        if tone == TONE1 or tone == TONE2:
            plt.figure(figsize=(10,3)) 
            plt.plot(fsweep,np.abs(Xdtft))
            plt.gca().set_ylabel('$|X(e^{j\omega})|$')
            plt.show()
        precisefmax = fsweep[np.argmax(np.abs(Xdtft[:Xdtft.size // 2]))]
        
        print(tone)
        print(fmax, precisefmax) 
    elif(tone >= 56):
        N = len(xall[tone])
        n = np.arange(0,N)
        X = np.fft.fft(xall[tone])
        FREQRANGE = 50
        FREQPOINTS = 200
        kall = np.arange(0,int(N/2) +1)
        Xmag = np.abs(X[kall])
        Xphase = np.angle(X[kall])
        f = kall / N * Fs
        if tone == TONE3:
            plt.plot(f[:f.size // 8],Xmag[:Xmag.size // 8])
            plt.gca().set_ylabel('$|X[k]|$')

        fmax = f[np.argmax(Xmag)]
        Xmax = np.max(Xmag)
        ffrom = fmax-FREQRANGE
        fto = fmax+FREQRANGE
        if tone == TONE3:
            plt.bar(ffrom,Xmax,fto-ffrom,linewidth=1,align='edge',color='orange',alpha=0.4) 
        fsweep = np.linspace(fmax-FREQRANGE, fmax+FREQRANGE,FREQPOINTS)

        A = np.zeros([FREQPOINTS, N],dtype=complex)   
        for k in np.arange(0,FREQPOINTS):
            A[k,:] = np.exp(-1j * 2 * np.pi * fsweep[k] / Fs * n) 
        Xdtft = np.matmul(A,xall[tone].T)
        if tone == TONE3:
            plt.figure(figsize=(10,3)) 
            plt.plot(fsweep,np.abs(Xdtft))
            plt.gca().set_ylabel('$|X(e^{j\omega})|$')
            plt.show()
        precisefmax = fsweep[np.argmax(np.abs(Xdtft))]
        print(tone)
        print(fmax, precisefmax)

        
```
